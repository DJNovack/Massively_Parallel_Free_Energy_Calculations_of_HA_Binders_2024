```python
def trinary_classifier(df,cutoff=1.2):
    classes = ['Destabilizing','Neutral','Stabilizing'] #'Neutral',
    columns = df.columns
    index = list(df.index)
    conditions = [df >= cutoff,(df > -cutoff) & (df < cutoff),df <= -cutoff]#(df > -cutoff) & (df < cutoff)
    df = np.select(conditions, classes)
    df = pd.DataFrame(df,columns=columns)
    df.index = index
    return df

def binary_classifier(df,cutoff=0.0):
    classes = ['Destabilizing','Stabilizing'] #'Neutral',
    columns = df.columns
    index = list(df.index)
    conditions = [df >= cutoff,df <= -cutoff]#(df > -cutoff) & (df < cutoff)
    df = np.select(conditions, classes)
    df = pd.DataFrame(df,columns=columns)
    df.index = index
    return df

def random_classifier(df,cutoff=1.2):
    classes = ['Destabilizing','Neutral','Stabilizing'] #'Neutral',
    p1 = df.sample(frac=1,axis=0)
    p2 = p1.sample(frac=1,axis=1)
    columns = df.columns
    index = list(df.index)
    conditions = [p2 >= cutoff,(p2 > -cutoff) & (p2 < cutoff),p2 <= -cutoff]#(df > -cutoff) & (df < cutoff)
    df = np.select(conditions, classes)
    df = pd.DataFrame(df,columns=columns)
    df.index = index
    return df

def matthews_curve(df, exp, cutoff):
    cls = trinary_classifier(df,cutoff=cutoff)
    co = matthews_corrcoef(exp.values.flatten(),cls.values.flatten())
    #co=pycm.ConfusionMatrix(exp.values.flatten(),cls.values.flatten()).overall_stat['Overall CEN']
    return co

def random_curve(df, exp, cutoff):
    cls = random_classifier(df,cutoff=cutoff)
    co = matthews_corrcoef(exp.values.flatten(),cls.values.flatten())
    #co=pycm.ConfusionMatrix(exp.values.flatten(),cls.values.flatten()).overall_stat['Overall CEN']
    return co
```

```python
ff_dic = np.load('../../data/mut_res_number.npy',allow_pickle=True).item()

fep_dic = {i:[] for i in ff_dic.keys()}
ros_dic = {i:[] for i in ff_dic.keys()}
random_dic = {i:[] for i in ff_dic.keys()}

dic = {'A8':A8,'A8_ros':A8_ros[A8.columns],'A13':A13,'A13_ros':A13_ros[A13.columns],'A18':A18,'A18_ros':A18_ros[A18.columns]}
exp_cutoff=1.8
fep_cutoff = 1.8
ros_cutoff=0.8
A8_exp_class = trinary_classifier(A8_exp[A8.columns],cutoff=exp_cutoff)
A13_exp_class = trinary_classifier(A13_exp[A13.columns],cutoff=exp_cutoff)
A18_exp_class= trinary_classifier(A18_exp[A18.columns],cutoff=exp_cutoff)
exp_dic = {'A8':A8_exp_class,'A13':A13_exp_class,'A18':A18_exp_class}

for key,val in dic.items():
    mnp = key.split('_')[0]
    exp = exp_dic[mnp]
    if 'ros' in key:
        classified = trinary_classifier(val,cutoff=ros_cutoff)
        mut_res = classified.index
        columns = classified.columns
        for i,col in enumerate(columns):
            wt_res = col[0]
            for mut in mut_res:
                res_type = f'{wt_res}2{mut}'
                e_val = exp[col][mut]
                v = classified[col][mut]
                try:
                    ros_dic[res_type].append([v,e_val])
                except:
                    continue
    else:
        classified = trinary_classifier(val,cutoff=fep_cutoff)
        mut_res = classified.index
        columns = classified.columns
        for i,col in enumerate(columns):
            wt_res = col[0]
            for mut in mut_res:
                if mut == wt_res:
                    continue
                res_type = f'{wt_res}2{mut}'
                print(res_type,end="\r")
                e_val = exp[col][mut]
                v = classified[col][mut]
                try:
                    fep_dic[res_type].append([v,e_val])
                except:
                    continue
                cats = ['Destabilizing','Neutral','Stabilizing']
                c_random = [cats[i] for i in np.random.randint(0,len(cats),size=500)]
                [random_dic[res_type].append([i,e_val]) for i in c_random] 
                
rev_ff_dic = {i:[] for i in np.unique(list(ff_dic.values()))}
for key,val in ff_dic.items():
    if fep_dic[key]:
        rev_ff_dic[val].append(key)
```

# Matrix of Charge Change by Number of Alchemical Transformations.

```python
negative = {'D':-1,'E':-1}
positive = {'K':1,'R':1}
neutral = {'A':0,'N':0,'C':0,'Q':0,'G':0,'I':0,'L':0,'M':0,'F':0,'P':0,'S':0,'T':0,'W':0,'Y':0,'V':0,'H':0}
all_aminos = neutral|positive|negative
r1,r2 = np.unique([i[0] for i in sum(list(rev_ff_dic.values()),[])]), np.unique([i[-1] for i in sum(list(rev_ff_dic.values()),[])])
df = pd.DataFrame(columns=r1,index=r2)
for key,item in ff_dic.items():
    try:
        res1 = key[0]
        res2 = key[-1]
        df[res1][res2]= all_aminos[res2] - all_aminos[res1]
    except:
        continue
df = df.fillna(0)
fig,ax = plt.subplots(dpi=300)
sns.heatmap(df,annot=True,ax = ax,cmap='coolwarm_r')
ax.xaxis.set_ticks_position('top')
ax.set_xlabel('WT Residue')
ax.xaxis.set_label_position('top') 
ax.set_title('Alchemical Transformation Charge Change')
ax.set_ylabel('Mut. Residue')
fig.tight_layout()
fig.savefig('charge_change_trans.pdf',dpi=300)
```
