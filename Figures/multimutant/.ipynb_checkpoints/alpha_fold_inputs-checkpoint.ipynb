```python
def make_mut_count_df(wildtype=None,x=None,FEP_df=None):
    residues = ['A', 'N', 'C', 'Q', 'G', 'I', 'L', 'M', 'F', 'P', 'S', 'T', 'W', 'Y', 'V', 'H', 'K', 'R', 'D', 'E']
    counts = {i:{res:0 for res in residues}  for i in wildtype}
    for row in x.values:
        seq,energy = row[0], row[1]
        for i,res in enumerate(seq):
            key = list(counts.keys())[i]
            counts[key][res]+=1
    #final_counts = {key1:{key2:((val2)/counts[key1][key1[0]]) for key2,val2 in val1.items()} for key1,val1 in counts.items()}
    final_counts = {key1:{key2:((val2)/sum(counts[key1].values())) for key2,val2 in val1.items()} for key1,val1 in counts.items()}
    for key1 in final_counts.keys():
        for key2 in final_counts[key1].keys():
            if key2==key1[0]:
                final_counts[key1][key2]= np.nan
    testdf = pd.DataFrame(final_counts)
    testdf=testdf[list(FEP_df.columns)]
    testdf = testdf.reindex(index = list(FEP_df.index))
    return testdf
```

```python
import numpy as np
from concurrent.futures import ProcessPoolExecutor
from tqdm import tqdm

def count_and_update(args):
    seq, target_positions, counts = args
    mutant_position_pairs = [(i, j) for i in target_positions for j in target_positions]

    for i, j in mutant_position_pairs:
        if seq[i] != seq[j]:  # Check if positions in the sequence differ
            counts[i, j] += 1

def calculate_conditional_probabilities(reference, strings,names=None, processes=8):
    target_positions = list(range(len(reference)))
    size = len(target_positions)
    counts = np.zeros((size, size), dtype=int)

    with ProcessPoolExecutor(max_workers=processes) as executor, tqdm(total=len(strings)) as pbar:
        try:
            for seq in strings:
                count_and_update((seq, target_positions, counts))
                pbar.update()
        except Exception as e:
            print(f"Exception in main: {e}")

    cond_probs = {}
    for i in target_positions:
        row_values = counts[i]
        if sum(row_values) != 0:
            cond_probs[i] = {j: row_values[j] / sum(row_values) for j in target_positions}
        else:
            cond_probs[i] = {j: 0 for j in target_positions if j != i}
    cond_probs = pd.DataFrame(cond_probs)
    cond_probs.index = names
    cond_probs.columns = names

    return cond_probs

# Example usage:
reference_string = ''.join([i[0] for i in wildtype_A18])
strings = list(xA18.sequence)

result = calculate_conditional_probabilities(reference_string, strings,names=wildtype_A18)

```

```python
def do_calc(cond_probs,indi):
    residues=cond_probs.columns
    copy = cond_probs.copy()
    combos = [(i, j) for i in residues for j in residues]
    for (r1,r2) in combos:
        if (indi[r1]*indi[r2]) == 0:
            copy[r1][r2]*= 0
        else:
            copy[r1][r2]/=(indi[r1]*indi[r2])
    return copy
```

```python
from itertools import product,permutations
import tqdm
import multiprocessing as mp
def make_combo_counts(wildtype=None,x=None,FEP_df=None):
    residues = ['A', 'N', 'C', 'Q', 'G', 'I', 'L', 'M', 'F', 'P', 'S', 'T', 'W', 'Y', 'V', 'H', 'K', 'R', 'D', 'E']
    numbers = [int(i[1:])-270 for i in FEP_df.columns]
    counts = np.zeros(shape=(len(numbers),len(numbers)))
    seqs = [row[0] for row in x[x['energy'] < -2].values]
    wildtype = ''.join([wildtype[i] for i in numbers])
    for seq in tqdm.tqdm(seqs):
        seq=''.join([seq[i] for i in numbers])
        idx = [j for j,i in enumerate(seq) if i!=wildtype[j]]
        temp_combos = list(permutations(idx,2))
        combos=[]
        for c in temp_combos:
            i,j = c
            if (j,i) not in combos:
                combos.append(c)
        for i,j in combos:
            if i!=j:
                counts[i,j]+=1
    return counts
    
```

```python
import tqdm
def make_combo_counts(seqs,wildtype=None):
    counts = np.zeros(shape=(len(wildtype),len(wildtype)))
    for seq in tqdm.tqdm(seqs):
        #seq=''.join([seq[i] for i in numbers])
        idx = [j for j,i in enumerate(seq) if i!=wildtype[j]]
        temp_combos = list(permutations(idx,2))
        combos=[]
        for c in temp_combos:
            i,j = c
            if (j,i) not in combos:
                combos.append(c)
        for i,j in combos:
            if i!=j:
                counts[i,j]+=1
    return counts
    
```

```python
from scipy.stats import linregress,t
from sklearn.metrics import mean_squared_log_error, mean_squared_error


fig,ax = plt.subplots(1,3,figsize=(12,4),dpi=300)
for n,name in enumerate(['A8','A13','A18']):
    entropy = np.loadtxt(f'{name}_mutation_entropy.txt')[:,1]
    covariance_matrix = covs[n]
    pair = np.zeros_like(covariance_matrix)
    for i in range(len(entropy)):
        for j in range(len(entropy)):
            if i != j:
                pair[i,j] += entropy[i]+entropy[j]
    # Flatten the matrices to 1D arrays
    entropy_matrix = pair
    xi, yi = entropy_matrix.flatten(), covariance_matrix.flatten()
    idx = np.argwhere(yi>0.0).flatten()
    xi,yi = xi[idx],yi[idx]
    # Assuming entropy_matrix and covariance_matrix are your matrices
    l=linregress(xi,np.log(yi))
    print("Correlation Coefficient:", l.rvalue, l.stderr)
    # Extract relevant information
    slope = round(l.slope,2)
    intercept = round(l.intercept,2)
    stderr_slope = l.stderr
    
    #RMSE={rmse} nats
    # Set confidence level (e.g., 95% confidence interval)
    confidence_level = 0.95
    
    # Degrees of freedom
    df = len(xi) - 2  # (number of data points - number of parameters)
    
    # Calculate t-value for the given confidence level and degrees of freedom
    t_value = t.ppf((1 + confidence_level) / 2, df)
    
    # Calculate margin of error for the slope
    margin_of_error_slope = t_value * stderr_slope
    
    # Calculate confidence intervals for slope
    lower_bound_slope = slope - margin_of_error_slope
    upper_bound_slope = slope + margin_of_error_slope
    
    # Calculate margin of error for the slope
    margin_of_error_slope = t_value * stderr_slope
    
    # Calculate confidence intervals for slope
    lower_bound_slope = slope - margin_of_error_slope
    upper_bound_slope = slope + margin_of_error_slope
    
    # Generate points for the linear fit line
    x_fit = np.linspace(min(xi), max(xi), 100)
    y_fit1 = slope * xi + intercept
    rmse=round(np.sqrt(mean_squared_error(yi, y_fit1)),2)

    print(rmse)
    y_fit = slope * x_fit + intercept
    # Generate points for the confidence intervals
    y_lower_bound = (slope - margin_of_error_slope) * x_fit + intercept
    y_upper_bound = (slope + margin_of_error_slope) * x_fit + intercept

    ax.flat[n].scatter(xi,np.log(yi),alpha=0.3)
    ax.flat[n].plot(x_fit,y_fit,c='k', label="Fitted Curve")
    ax.flat[n].set_xlabel('$\sum$ Entropy (nats)')
    ax.flat[n].set_ylabel('ln(CoV)')
    ax.flat[n].set_title(f'{name}')
    ax.flat[n].text(np.min(xi),np.max(np.log(yi))-0.5,f'ln(CoV)={slope}S+{intercept}\nR-value: {l.rvalue:.2f}', fontsize=12)
    ax.flat[n].fill_between(x_fit, y_lower_bound, y_upper_bound, color='gray', alpha=0.3, label='95% Confidence Intervals')

# Add labels and legend
fig.tight_layout()
#fig.savefig('log_cov_entropy.png',dpi=300)
```

```python
from scipy.stats import entropy
from scipy.integrate import quad
beta = 0.593**-1

def shannon_entropy(df):
    # Ensure the DataFrame is not empty
    if df.empty:
        raise ValueError("DataFrame is empty")

    # Calculate probabilities for each column
    probabilities = df.apply(lambda col: np.exp(-beta*col) / np.sum(np.exp(-beta*col)))

    # Calculate Shannon entropy using scipy's entropy function
    entropy_values = [-np.nansum([p*np.log(p) for p in probabilities[col]]) for col in probabilities.columns]

    # Create a new DataFrame with entropy values and column names
    result_df = pd.DataFrame({"Residue": df.columns, "Shannon_Entropy": entropy_values})

    return result_df


def bhattacharyya_distances(position_wise_entropy_p, position_wise_entropy_q):
    # Calculate the Bhattacharyya coefficient
    bc = np.sqrt(position_wise_entropy_p * position_wise_entropy_q)

    # Bhattacharyya distance
    bhattacharyya_distance = -np.log(bc.sum())

    return round(bhattacharyya_distance,1)

def make_ent_df(file,wildtype):
    test=np.loadtxt(file)
    t=pd.DataFrame(test[:,1])
    t.index=([j for j in wildtype if int(j[1:]) in test[:,0].astype(int)])
    return t.T

#test = shannon_entropy(clipped_df)
```
