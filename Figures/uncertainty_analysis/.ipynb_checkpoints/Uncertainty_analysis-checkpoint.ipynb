```python
import os,sys
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
A8_err = pd.read_pickle('../../data/EE/A8_variance.pkl')
A13_err = pd.read_pickle('../../data/EE/A13_variance.pkl')
A18_err = pd.read_pickle('../../data/EE/A18_variance.pkl')
```

# Plot Histogram of Uncertainties

```python
A8_v = A8_err.values.flatten()
A13_v = A13_err.values.flatten()
A18_v = A18_err.values.flatten()
values = np.concatenate([A8_v,A13_v,A18_v])
mean = np.mean(values)
fig,ax = plt.subplots(dpi=300)
sns.histplot(values,ax=ax,bins=50)
ax.axvline(mean,0,300,c='k',linewidth=0.5)
ax.text(8,200,f'$\mu$={round(mean,2)} kcal/mol')
ax.set_ylim(0,260)
ax.set_ylabel('Counts')
ax.set_xlabel('$\sigma$ (kcal/mol)')
fig.savefig('sigma_hist.png',dpi=300)
```
